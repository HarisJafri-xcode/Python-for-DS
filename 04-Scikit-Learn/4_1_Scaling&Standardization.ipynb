# Importing the Data Science & Machine Learning Libraries

Many a times, it is important to scale or standardize the features of a DataFrame so that our Machines may perform Machine Learning Techniques correctly.

```python
import pandas as pd # Python Module for Data Frame Handling
import numpy as np # NumPy Arrays are primarily used for Machine Learning Libraries like Scikit-Learn
from sklearn.preprocessing import MinMaxScaler , StandardScaler # Machine Learning Library
```

# Storing Both Transformers into a Variable



```python
minmaxscaler = MinMaxScaler() # Instantiating the MinMaxScaler
```

```python
standardscaler = StandardScaler() # Instantiating the StandardScaler
```

# Working with Marital Data Frame

```python
numerical_features = ['Age (years)', 'Income ($)']
```

Copying the Original DataFrame into a DataFrame where we are actually going to perform MinMax Scaling and Standardization.

```python
df_marital_minmaxscaled = df_marital.copy()
```

```python
df_marital_standardscaled = df_marital.copy()
```

## Min Max Scaling

```python
df_marital_minmaxscaled[numerical_features] = minmaxscaler.fit_transform(df_marital_minmaxscaled[numerical_features])
```

```python
df_marital_minmaxscaled
```

## Standard Scaler

```python
df_marital_standardscaled[numerical_features] = standardscaler.fit_transform(df_marital_standardscaled[numerical_features])
```

```python
df_marital_standardscaled
```

```python
print(round(np.mean(df_marital_standardscaled['Age (years)']))) # Checking whether Mean is 0
```

```python
print(round(np.std(df_marital_standardscaled['Age (years)']))) # Checking whether St. Dev is 1
```
