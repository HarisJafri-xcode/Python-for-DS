# zip() and enumerate() in Python

Two powerful helpers for loops and data pairing:
- `zip()` → combine multiple iterables element by element
- `enumerate()` → get index + value while looping

## 1. zip() – Pairing Iterables

`zip()` combines multiple iterables (lists, tuples, etc.) element by element into tuples.

- Stops at the shortest iterable
- Returns an iterator of tuples

```python
names = ["Alice", "Bob", "Charlie"]
scores = (85, 35, 70)

for name, score in zip(names, scores):
    print(name, "made", score)
```

```python
students = ["Ali", "Sara", "Omar"]
math = [90, 80, 70]
physics = [88, 75, 72]

for name, m, p in zip(students, math, physics):
    print(name, "-> Math:", m, "Physics:", p)
```

### zip() to list
Convert a zip object to a list to see all pairs at once.

```python
pairs = list(zip(names, scores))
pairs
```

## 2. enumerate() – Index + Value

`enumerate()` gives you both **index** and **value** when looping.

Instead of manual counters, use `enumerate()`.

```python
fruits = ['apple', 'banana', 'cherry']

for index, fruit in enumerate(fruits):
    print(index, fruit)
```

### Custom starting index

Use `start=` to control where counting begins (e.g., 1 instead of 0).

```python
for roll, fruit in enumerate(fruits, start=1):
    print("Roll #", roll, "->", fruit)
```

## 3. Combining zip() and enumerate()

You can mix them to get index + paired data in one go.

```python
names = ["Ali", "Sara", "Omar"]
scores = [90, 80, 70]

for idx, (name, score) in enumerate(zip(names, scores), start=1):
    print(f"{idx}. {name} scored {score}")
```
