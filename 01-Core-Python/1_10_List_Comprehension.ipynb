# List Comprehension in Python

List comprehensions are a concise way to create lists using a single line.

Pattern:
```python
[ expression for item in iterable if condition ]
```

## 1. Basic List Comprehension

Create `[0, 1, 2, 3, 4]` quickly.

````python
normal = [i for i in range(5)]
print(normal)
````

### Squares and cubes

Modify `i` in the expression to get other patterns.

````python
squares = [i**2 for i in range(5)]
cubes = [i**3 for i in range(5)]
print("squares:", squares)
print("cubes:", cubes)
````

## 2. List Comprehension with Condition

Filter elements by adding an `if` at the end.

````python
cubes_c = [i**3 for i in range(5) if i**3 < 50]
print(cubes_c)
````

Example: numbers from 0 to 19 divisible by 3.

````python
x = [i for i in range(20) if (i % 3) == 0]
print(x)
````

## 3. Combining Operations and Conditions

Example from slide:
`[i*i*2 for i in range(3)]  -> [0, 2, 8]`

````python
y = [i * i * 2 for i in range(3)]
print(y)
````
